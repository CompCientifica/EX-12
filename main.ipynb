```python
import simpy
def chegada_navio(env, porto):
    navio_id = 1
    while True:
        yield env.timeout(1)  # Intervalo de cada navio chegando
        print(f'Navio {navio_id} chegou ao porto em {env.now}')
        env.process(atracar_navio(env, f'Navio {navio_id}', porto))
        navio_id += 1
def atracar_navio(env, nome, porto):
    with porto.request() as requisicao:
        yield requisicao
        print(f'{nome} atracou no porto em {env.now}')
        yield env.timeout(5)  # Tempo médio do navio no porto
        print(f'{nome} partiu do porto em {env.now}')
# Configurações da simulação
env = simpy.Environment()
porto = simpy.Resource(env, capacity=5)  # Capacidade do Porto
# Inicie a simulação
env.process(chegada_navio(env, porto))
env.run(until=50)  # Ate dar 50 unidades de tempo
```
